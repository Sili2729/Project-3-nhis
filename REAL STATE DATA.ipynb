```python
import pandas as pd

# 1. Overview of the dataset
print("Initial Shape:", df.shape)
print("Data Types:\n", df.dtypes)
print("Missing Values:\n", df.isnull().sum())

# 2. Drop completely empty rows (if any)
df.dropna(how='all', inplace=True)

# 3. Handle missing values

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# 4. Remove duplicate entries
df.drop_duplicates(inplace=True)

# 5. Remove or handle anomalies
if 'Price' in df.columns and 'Area' in df.columns:
    df = df[(df['Price'] > 0) & (df['Area'] > 0)]

# 6. Reset index
df.reset_index(drop=True, inplace=True)

# 7. Confirm cleaning
print("Cleaned Shape:", df.shape)
print("Any Missing Values Left?\n", df.isnull().sum())
```

```python
file_path = (r"C:\Users\Hitansh\Desktop\PROJECT 3\housing_data.csv")
df = pd.read_csv(file_path)

# Data cleaning
df.dropna(how='all', inplace=True)

# Fill missing numerical values with median
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove anomalies (e.g., Price <= 0 or Area <= 0 if those columns exist)
for col in ['Price', 'Area']:
    if col in df.columns:
        df = df[df[col] > 0]

# Reset index
df.reset_index(drop=True, inplace=True)

clean_file_path = (r"C:\Users\Hitansh\Desktop\PROJECT 3\housing_data cleaned.csv")
df.to_csv(clean_file_path, index=False)
```

```python
import pandas as pd
import calendar

# Load the cleaned dataset
df = pd.read_csv(r"C:\Users\Hitansh\Desktop\PROJECT 3\housing_data cleaned.csv")

# Step 1: Convert month names (e.g., 'Feb') to numbers
month_map = {v: k for k, v in enumerate(calendar.month_abbr) if v}
df['MoSold'] = df['MoSold'].map(month_map)

# Step 2: Create a proper SaleDate column
df['SaleDate'] = pd.to_datetime(df['YrSold'].astype(str) + '-' + df['MoSold'].astype(str), format='%Y-%m')

# Step 3: Extract full month name (e.g., "February") into a new column
df['MonthName'] = df['SaleDate'].dt.strftime('%B')

# Step 4: Show a preview
print(df[['SaleDate', 'MonthName']].head(10))
```

```python

df['MonthYearLabel'] = df['SaleDate'].dt.strftime('%b-%Y')

# Group by MonthYear label and calculate average sale price
month_year_avg = df.groupby('MonthYearLabel')['SalePrice'].mean().reset_index()

# To ensure chronological order, also extract datetime
month_year_avg['SaleDate'] = pd.to_datetime(month_year_avg['MonthYearLabel'], format='%b-%Y')
month_year_avg.sort_values('SaleDate', inplace=True)

# Plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=month_year_avg, x='MonthYearLabel', y='SalePrice', marker='o', color='orange')
plt.title("📅 Average Sale Price by Month-Year")
plt.xlabel("Month-Year")
plt.ylabel("Average Sale Price")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Extract year
df['Year'] = df['SaleDate'].dt.year

# Group by year and average price
year_avg = df.groupby('Year')['SalePrice'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=year_avg, x='Year', y='SalePrice', palette='viridis')
plt.title("📈 Average Sale Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Sale Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\Hitansh\Desktop\PROJECT 3\housing_data cleaned.csv")

# List of important features
important_features = [
    'SalePrice',
    'GrLivArea',
    'TotalBsmtSF',
    'GarageArea',
    'GarageCars',
    'OverallQual',
    'YearBuilt',
    'FullBath',
    '1stFlrSF'
]


corr_matrix = df[important_features].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.1)

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    square=True,
    annot_kws={"size": 10}
)

plt.title("Correlation Heatmap of Important Features", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
